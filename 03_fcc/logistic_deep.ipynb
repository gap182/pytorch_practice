```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from torchvision.datasets import MNIST 
import torchvision
from torch.utils.data import DataLoader
import torch.nn.functional as F

%matplotlib inline
```

```python
data_train = MNIST('data/', download=True, train=True, transform=torchvision.transforms.ToTensor())
data_test = MNIST('data/', download=True, train=False, transform=torchvision.transforms.ToTensor())

print('Shape of train/test MNIST data is: {}/{}'.format(len(data_train), len(data_test)))
```

```python
image, label = data_train[0]

plt.imshow(image[0], cmap='gray')
print(label)
```

```python
train_data = DataLoader(data_train, batch_size=50, shuffle=True)
test_data = DataLoader(data_test, batch_size=50, shuffle=True)
```

```python
class MNISTmodel(nn.Module):
    def __init__(self,n_input,n_classes,H):
        super(MNISTmodel,self).__init__()
        self.linear1 = nn.Linear(n_input,H)
        self.linear2 = nn.Linear(H,n_classes)
        
    def forward(self,X):
        X = X.view(-1,784)
#         X = F.relu(self.linear1(X))
        X = self.linear1(X)
        return X
```

```python
def fit(model, loss_fun, data, opt, epochs, lr):
    for i in range(epochs+1):
        for xb, yb in data:
            z = model(xb)
            loss = loss_fun(yb,z)
            
            loss.backward()
            opt.step()
            opt.zero_grad()
            
            if i%50 == 0:
                print(f'Epoch [{i}/{epochs}], Loss: {loss}')
    
```

```python
input_size = 784
num_classes = 10
hidden = 10

model = MNISTmodel(input_size, num_classes, hidden)
lr = 0.001
epochs = 500

loss_fn = F.cross_entropy
opt = torch.optim.SGD(model.parameters(), lr=lr)
```

```python
fit(model, F.cross_entropy, train_data, opt, epochs, lr)
```
